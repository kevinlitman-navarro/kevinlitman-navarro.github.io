```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
response = requests.get('https://www.basketball-reference.com/leagues/NBA_wins.html').content
soup_doc = BeautifulSoup(response, 'html.parser')
```

```python
table = soup_doc.table
```

```python
seasons = table.find_all('tr')[1:-1]
```

```python
seasons[0].text[1:5]
```

```python
seasons[0].find_all('td')[0].text[:4]
```

```python
seasons[-1].find_all('td')[2].text
```

```python
rows = []

for season in seasons:
    try:
        continuity_dict = {}
        continuity_dict['season'] = season.find_all('td')[0].text[:4]
        continuity_dict['ATL'] = season.find_all('td')[2].text
        continuity_dict['BOS'] = season.find_all('td')[3].text
        continuity_dict['CHA'] = season.find_all('td')[6].text
        continuity_dict['CHI'] = season.find_all('td')[5].text
        continuity_dict['CLE'] = season.find_all('td')[7].text
        continuity_dict['DAL'] = season.find_all('td')[8].text
        continuity_dict['DEN'] = season.find_all('td')[9].text
        continuity_dict['DET'] = season.find_all('td')[10].text
        continuity_dict['GSW'] = season.find_all('td')[11].text
        continuity_dict['HOU'] = season.find_all('td')[12].text
        continuity_dict['IND'] = season.find_all('td')[13].text
        continuity_dict['LAC'] = season.find_all('td')[14].text
        continuity_dict['LAL'] = season.find_all('td')[15].text
        continuity_dict['MEM'] = season.find_all('td')[16].text
        continuity_dict['MIA'] = season.find_all('td')[17].text
        continuity_dict['MIL'] = season.find_all('td')[18].text
        continuity_dict['MIN'] = season.find_all('td')[19].text
        continuity_dict['NJN'] = season.find_all('td')[4].text
        continuity_dict['NOH'] = season.find_all('td')[20].text
        continuity_dict['NYK'] = season.find_all('td')[21].text
        continuity_dict['OKC'] = season.find_all('td')[22].text
        continuity_dict['ORL'] = season.find_all('td')[23].text
        continuity_dict['PHI'] = season.find_all('td')[24].text
        continuity_dict['PHO'] = season.find_all('td')[25].text
        continuity_dict['POR'] = season.find_all('td')[26].text
        continuity_dict['SAC'] = season.find_all('td')[27].text
        continuity_dict['SAS'] = season.find_all('td')[28].text
        continuity_dict['TOR'] = season.find_all('td')[29].text
        continuity_dict['UTA'] = season.find_all('td')[30].text
        continuity_dict['WAS'] = season.find_all('td')[31].text
        rows.append(continuity_dict)
    except:
        pass
df = pd.DataFrame(rows)
```

```python
df.replace({'': np.nan}, regex=True, inplace=True)
```

```python
df.to_csv('historical-win-total.csv', index=False)
```
